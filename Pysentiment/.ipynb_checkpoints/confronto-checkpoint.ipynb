```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as dates
from sklearn import preprocessing


indexes = pd.read_csv('indexesT.csv')

indexes.head()
```

```python
#Normalizing indexes

values = {'boll_ub': 138, 'boll_lb': 137}
indexes = indexes.fillna(value=values)
indexes = indexes.fillna(0)

x = indexes[['high','boll_ub','boll_lb']].values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
indexes[['high','boll_ub','boll_lb']] = pd.DataFrame(x_scaled)

indexes[['high','boll_ub','boll_lb']] -= [0.62,0.62,0.62]


plt.figure(figsize=(20,10))
plt.plot(indexes['high'].tolist(),'b')
#Media mobile anche sugli indici
indexes[['high','boll_ub','boll_lb']] = indexes[['high','boll_ub','boll_lb']].rolling(window=60).mean()

plt.plot(indexes['high'].tolist(),'r')



plt.show()
```

```python
sentimentTextBlob = pd.read_csv('sentiment20minutes.csv')
meanSentiment = sentimentTextBlob['sentiment'].mean()
meanSentimentWeighted = sentimentTextBlob['sentimentWeighted'].mean()

sentimentTextBlob['sentiment'] -= meanSentiment
sentimentTextBlob['sentimentWeighted'] -= meanSentimentWeighted




sentimentTextBlob.head()
```

```python
# Dataset allignment in time
indexes = indexes[indexes['date'] >= sentimentTextBlob['initTime'][0]]
#sentimentTextBlob = sentimentTextBlob[sentimentTextBlob['initTime'] < indexes['Unnamed: 0'][len(indexes)-1]]

indexes.head()
```

```python
print(len(sentimentTextBlob['initTime'].tolist()))
print(len(indexes['date'].tolist()))
```

```python
plt.figure(figsize=(20,10))
plt.plot(indexes['high'].tolist(),'b',sentimentTextBlob['sentiment'].tolist(),'r',)
plt.title('Only Sentiment')
plt.show()
```

```python
plt.figure(figsize=(20,10))
#plt.plot(indexes['high'],'b',indexes['boll_ub'],'r',indexes['boll_lb'],'g')
plt.plot(indexes['high'].tolist(),'b',sentimentTextBlob['sentimentWeighted'].tolist(),'r',)
plt.title('Weighted Sentiment')
plt.show()
```

```python
sentiment5 = sentimentTextBlob['sentiment'].rolling(window=5).mean()
sentiment15 = sentimentTextBlob['sentiment'].rolling(window=15).mean()
sentiment30 = sentimentTextBlob['sentiment'].rolling(window=30).mean()
sentiment50 = sentimentTextBlob['sentiment'].rolling(window=50).mean()
sentiment100 = sentimentTextBlob['sentiment'].rolling(window=100).mean()


plt.figure(figsize=(20,10))
#plt.plot(indexes['high'],'b',indexes['boll_ub'],'r',indexes['boll_lb'],'g')
plt.plot(indexes['high'].tolist(),'b',sentiment100,'y',sentiment50,'c',np.zeros(len(sentiment30)),'r')
plt.title('Comparison between moving averages')
plt.show()


sentimentTextBlob['sentiment'] = sentimentTextBlob['sentiment'].rolling(window=35).mean()
sentimentTextBlob['sentimentWeighted'] = sentimentTextBlob['sentimentWeighted'].rolling(window=35).mean()
```

```python

plt.figure(figsize=(20,10))
#plt.plot(indexes['high'],'b',indexes['boll_ub'],'r',indexes['boll_lb'],'g')
plt.plot(indexes['high'].tolist(),'b',sentimentTextBlob['sentiment'].tolist(),'r',)
plt.title('Only Sentiment Moving Average')
plt.show()
```

```python
plt.figure(figsize=(20,10))
#plt.plot(indexes['high'],'b',indexes['boll_ub'],'r',indexes['boll_lb'],'g')
plt.plot(indexes['high'].tolist(),'b',sentimentTextBlob['sentimentWeighted'].tolist(),'r',)
plt.title('Sentiment Weighted Moving Average')
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.plot(sentimentTextBlob['sentimentWeighted'].tolist(),'b',sentimentTextBlob['sentiment'].tolist(),'r',)
plt.show()
```

```python
#Visto che sono simili d'ora in poi considero solo sentiment normale
 #Derivata

slopeStock = pd.Series(np.gradient(indexes['high']))
accuracy = [a*b*1000 for a,b in zip(slopeStock,sentiment100)]


plt.figure(figsize=(20,10))
plt.plot(indexes['high'].tolist(),'b',sentiment100,'y',accuracy,'g',np.zeros(len(sentiment30)),'r')
plt.show()
```
