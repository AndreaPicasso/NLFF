```python
import nltk
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
from gensim.models import KeyedVectors
```

Our features are composed of the averaged word2vec vector for the words that compose concepts in the selected news (concepts that aren't stop words)

Prendi Concepts da SenticNet folder


```python
stop_words = set(stopwords.words('english'))
stop_words = stop_words - set(('up','down','above','more','most','over','under','below'))
```

```python
aapl = pd.read_csv('/home/simone/Desktop/NLFF/SenticNet/conceptsDataset/summary/AAPL.csv')
features_names = ['DATE','polarity','attention','pleasantness','aptitude','sensitivity','stemmed_polarity','stemmed_attention',
                  'stemmed_pleasantness','stemmed_aptitude','stemmed_sensitivity','concepts','stemmed_concepts']
aapl = aapl[features_names]
aapl.head()
```

```python
w2v = KeyedVectors.load_word2vec_format('/home/simone/Desktop/GoogleNews-vectors-negative300.bin', binary=True)
# w2v['stock'] 
```

```python
vecs = list()
for index, row in aapl.iterrows():
    if index%500 == 0:
        print(index)
    tempVec = np.zeros([300,])
    concepts =''
    concepts += row['concepts']+' ' if not str(row['concepts'])=='nan' else ''
    concepts += row['stemmed_concepts'] if not str(row['stemmed_concepts'])=='nan'  else ''
    concepts = concepts.split(' ')
    concepts = set(concepts) - stop_words
    
    num_concept_presents = 0
    present_concepts = list()
    for concept in concepts:
        if concept in w2v:
            present_concepts.append(concept)
            num_concept_presents +=1
            tempVec += w2v[concept]
    if(num_concept_presents > 0):
        tempVec = tempVec/num_concept_presents
    print(present_concepts)
    vecs.append(tempVec)
```

```python
newsW2v = pd.DataFrame(data=np.asarray(vecs))
newsW2v[newsW2v.isnull().any(axis=1)]
```

```python
newsW2v['DATE'] = aapl['DATE']
```
