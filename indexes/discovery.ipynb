```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tickers = ['AAPL','AMZN','GOOGL','MSFT','FB','INTC','CSCO','CMCSA','NVDA','NFLX']       
```

```python
## The 'Close' field is actually the 'adjusted close', same for the open I think
for tick in tickers:
    price = pd.read_csv('indexes'+str(tick)+'.csv')
    m = price['close'].min()
    M = price['close'].max()
    plt.figure(figsize=(15,5))
    plt.plot((price['close'] - m)/(M-m), label = tick)
    m = price['open'].min()
    M = price['open'].max()
    plt.plot((price['open'] - m)/(M-m), label = tick)
    
    plt.legend()
    plt.show()
    
```
