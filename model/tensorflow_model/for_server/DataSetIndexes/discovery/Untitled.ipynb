```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
stock = pd.read_csv('../indexesAAPL.csv')
```

```python
plt.figure(figsize=(20,10))
plt.plot(range(len(stock['open'])),stock['open'],'r',stock['close'],'b')

```

```python
# Remove mean
ma = stock['close'].max()
mi = stock['close'].min()
for index, row in stock.iterrows():
    stock.at[index, 'close'] = 2*(row['close'] - mi)/(ma - mi) -1

```

```python
MOMENTUM_WINDOW = 100


y = list()
for i in range(0,MOMENTUM_WINDOW):
    y.append(-1)
for i in range(MOMENTUM_WINDOW,stock.shape[0]):
    y.append(stock['close'][i] - stock['close'][i-MOMENTUM_WINDOW])
y = np.sign(y)
plt.figure(figsize=(20,10))
plt.plot(range(len(stock['close'])),y,'r',stock['close'],'b')
```
