```python
# coding=utf-8

import numpy as np
import pandas as pd
import json
import ast
import requests
```

```python
news = pd.read_csv("/home/simone/Desktop/NLFF/intrinioDataset/preprocessing/preprocessed/AAPL.csv")
news.head()
```

```python
out = pd.DataFrame(columns = ['date', 'SUMMARY', 'concepts'])
with open ("/home/simone/Desktop/NLFF/FilippoDataSet/list_OL_FINANCE.txt", "r") as myfile:
            data=myfile.read()
data = ast.literal_eval(data) 
filConcepts = json.loads(json.dumps(data))
```

```python
URL = "http://sentic.net/api/7553HJ/"
j = 0
for i, row in news.iterrows():
    concepts = list()
    try:
        sentence = news["SUMMARY"].at[i]
        response = requests.get(URL+sentence)
        #data = json.load(response)
        data = response.json()
        if "aspects" in data:
            concepts.append(data["aspects"].items())
        words = sentence.split()
        concepts.append(set(words).union(filConcepts.keys()))
        print(str(i)+"/"+str(len(news.index)))
    except:
        print("ERROR: "+sentence)
    out['date'].at[j] = news["Unnamed: 0"].at[i]
    out['SUMMARY'].at[j] = news["SUMMARY"].at[i]
    out['concepts'].at[j] = concepts
    j +=1
    
    
```

```python
out.to_csv("AAPL_concepts.csv")
```
