```python
import numpy as np
import pandas as pd
```

```python
LMD_dict = pd.read_excel('LoughranMcDonald_MasterDictionary_2014.xlsx')
LMD_dict.head()
```

```python
Negative = []
Positive = []
Uncertainty = []
Litigious = []
Constraining = []
Superfluous = []
Interesting = []

for index, row in LMD_dict.iterrows():
    if(row['Negative'] != 0):
        Negative.append(row['Word'])
    if(row['Positive'] != 0):
        Positive.append(row['Word'])
    if(row['Uncertainty'] != 0):
        Uncertainty.append(row['Word'])
    if(row['Litigious'] != 0):
        Litigious.append(row['Word'])
    if(row['Constraining'] != 0):
        Constraining.append(row['Word'])
    if(row['Superfluous'] != 0):
        Superfluous.append(row['Word'])
    if(row['Interesting'] != 0):
        Interesting.append(row['Word'])
```

```python
Negative = pd.DataFrame({'NEGATIVE':Negative})
Positive = pd.DataFrame({'POSITIVE':Positive})
Uncertainty = pd.DataFrame({'UNCERTAINTY':Uncertainty})
Litigious = pd.DataFrame({'LITIGIOUS':Litigious})
Constraining = pd.DataFrame({'CONSTRAINING':Constraining})
Superfluous = pd.DataFrame({'SUPERFLUOUS':Superfluous})
Interesting = pd.DataFrame({'INTERESTING':Interesting})

myDict = pd.concat([Negative,Positive,Uncertainty,Litigious,Constraining,Superfluous,Interesting], ignore_index=True, axis=1)
myDict.columns = ['NEGATIVE', 'POSITIVE','UNCERTAINTY', 'LITIGIOUS','CONSTRAINING', 'SUPERFLUOUS','INTERESTING' ]

myDict.head()
```

```python
myDict.to_csv('LoughranMcDonald_MyDictionary2.csv')
```
